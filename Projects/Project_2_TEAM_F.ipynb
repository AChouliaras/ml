# Linear Algebra

\textbf{Vectors and Vector Spaces} 

Vector spaces are the basic setting in which linear algebra happens. A \textbf{vector space V} is a set (the elements of which are called vectors) on which two operations are defined: vectors can be added together, and vectors can be multiplied by real numbers called scalars.

A vector space V should satisfy the following:

\begin{itemize}
  \item There exists an additive identity (written 0) in V such that $ x + 0 = x$  for all $x \in V$
  \item For each $x\in V$, there exists an additive inverse (written −x) such that $x + (−x) = 0$
  \item There exists a multiplicative identity (written 1) in R such that $ 1x = x $ for all $x \in V$
  \item Commutativity: $x + y = y + x$ for all $x,y \in V$
  \item Associativity: $ (x + y) + z = x + (y + z)$ and $α(βx) = (αβ)x $ for all $x,y,z\in V$ and $α,β \in R$
  \item Distributivity: $ α(x + y) = αx + αy$ and $(α + β)x = αx + βx $ for all $x,y\in V$ and $α,β\in R$
\end{itemize}

A set of $n$ vectors $v_1,...,v_n$ is \textbf{linearly indepedent} if: <br>
$$a_1v_1 + ... + a_nv_n = 0$$ , only in the case when:
$a_1 = ... = a_n = 0$ <br><br>

It is usefull and easy to think of vectors as $Nx1$ matrices: 
$\begin{bmatrix}x_1\\ \vdots \\ x_n\end{bmatrix}$

<br>

Addition and scalar multiplication are shown bellow:

$\begin{bmatrix}x_1\\ \vdots \\ x_n\end{bmatrix}$ + $\begin{bmatrix}y_1\\ \vdots \\ y_n\end{bmatrix}$ = $\begin{bmatrix}x_1 + y_1\\ \vdots \\x_n + y_n\end{bmatrix}$

In the following block of Python code, we examine some examples of the six conditions mentioned above.

```python
import numpy as np
from numpy import linalg
import math
import matplotlib.pyplot as plt

# two vectors 
x = np.array([1, 5, 7])
y = np.array([5, 7, 14])
z = np.array([5, 9, 4])
zero = np.array([0, 0, 0])
minus_x = -x

print("x = ")
print(x)
print("\n")

print("y = ")
print(y)
print("\n")

# There exists an additive identit
print( "The statement x + 0 = x is " + str((x + zero).all() == (x).all()))

# For each x, there exists an additive inverse
print( "The statement x + (-x) = 0 is " + str((x + minus_x).all() == zero.all()))

# There exists a multiplicative identity
print( "The statement 1*x = x is " + str((1*x).all() == x.all()))

# Commutativity:
print( "The statement x + y = y + x is " + str((x + y).all() == (y + x).all()))

# Associativity
print( "The statement (x + y) + z = x + (y + z) is " + str(((x + y) + z).all() == (x + (y + z)).all()))

#Distributivity: $ α(x + y) = αx + αy$ and $(α + β)x = αx + βx $
print( "The statement α(x + y) = αx + αy is " + str((5*(x + y)).all() == (5*x + 5*y).all()))
print( "The statement (α + β)x = αx + βx is " + str(((5 + 10)*x).all() == (5*x + 10*x).all()))
```

\textbf{Linear Maps}

A linear map or a linear transformation, is a function $T: V\rightarrow W$ where $V$ and $W$ are vector spaces, that satisfies:

\begin{itemize}
    \item $T(x + y) = Tx + Ty$, for all $ x,y\in V$
    \item $T(ax) = aTx$, for all $x\in V, a\in R$
\end{itemize}

Some basic examples of linear transformations may be:

\begin{itemize}
    \item \textbf{Expansion Matrix}: $\begin{bmatrix}c & 0 \\0 & c\end{bmatrix}$ 
    \item \textbf{Rotation}: $\begin{bmatrix}cos(θ) & -sin(θ) \\sin(θ) & cos(θ)\end{bmatrix}$
    \item \textbf{Reflection}: $\begin{bmatrix} 2cos(θ)^2 - 1 & 2cos(θ)sin(θ) \\2cos(θ)sin(θ) & 2sin(θ)^2 - 1\end{bmatrix}$
    \item \textbf{Projection}: $\begin{bmatrix} cos(θ)^2 & cos(θ)sin(θ) \\cos(θ)sin(θ) & sin(θ)^2\end{bmatrix}$ 
\end{itemize}

In the following block, we present an example of each linear map:

\textbf{Transpose of a matrix}

Transpose $A^T \in R^{n×m}$ is given by $(A^T)ij=Aji$ for each $(i,j)$. <br>

The transpose has several nice algebraic properties that can be easily verified from the definition:

\begin{itemize}
    \item $(A^T)^T = A$
    \item $(A + B)^T = A^T + B^T$
    \item $(aA)^T = aA^T$
    \item $(AB)^T = B^TA^T$
\end{itemize}

Bellow, we examine the transpose of a matrix A as well as the properties mentioned above:

\textbf{Normed Spaces}

A norm on a real vector space V is a function $||\dots||: V \rightarrow R$, that satisfies:

\begin{itemize}
    \item $||x|| \geq 0$, with equality only when $x = 0$
    \item $||ax|| = |a|||x||$
    \item $||x + y|| \leq ||x|| + ||y||$
 \end{itemize}
 
Any norm on $V$ induces a distance metric on $V$:

$$d(x,y) = ||x - y||$$

Following, we show some norms we will use:

$$||x||_1 = \sum^{n}_{i = 1}|x_i|$$

$$||x||_2 = \sqrt{\sum^{n}_{i = 1}x_i^2}$$

$$||x||_p = (\sum^{n}_{i = 1}x_i^p)^{1/p},     (p \geq 1)$$

$$||x||_{\infty} = max_{1\leq i\leq n}|x_i|$$

In the following block, we use numpy(and especially linalg package) to calculate some of the norms mentioned above:

```python
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])

print(linalg.norm(x, 1))

print(linalg.norm(x, 2))

print(linalg.norm(x, np.inf))

print(linalg.norm(x, 4))
```

\textbf{Inner product spaces} 

An inner product on a real vector space V is a function $〈·,·〉: V×V→R$ satisfying:

\begin{itemize}
    \item $\langle x,x \rangle \geq 0$, with equality if and only if $x = 0$
    \item Linearity in the first slot: $ \langle x + y,z \rangle = \langle x,z \rangle + \langle y,z \rangle$ and $ \langle ax,y \rangle = a\langle x,y \rangle $
    \item $ \langle x,y \rangle = a\langle y,x \rangle $
\end{itemize}

for all $x,y,z \in V$ and $a \in R$

Any inner product on V induces a norm on V:

$$ ||x|| = \sqrt{\langle x,x \rangle} $$

Two vectors $x$ and $y$ are \textbf{orthogonal} if $\langle x,y \rangle = 0$ and we write $x \perp y$. Additionally, if $||x|| = ||y|| = 1$, we say that the vectors are \textbf{orthonormal}. 

The standard inner product on $R^n$ is given by:

$$\langle x,y \rangle = x^Ty$$

```python
x = np.array([2, 6, 4])
y = np.array([5, 9, 5])
zero = np.array([0, 0, 0])

a = np.array([2, 4])
b = np.array([-1, 0.5])

c = np.array([1, 0])
d = np.array([0, 1])


print("inner product of x: ", np.inner(x,x))
print("inner product of y: ", np.inner(y,y))
print("inner product of zero: ", np.inner(zero,zero))

print("inner product of xy: ", np.inner(x,y))

print("\ninner product of ab: ", np.inner(a,b))
print("Vectors a and b are orthogonal")

print("\ninner product of cd: ", np.inner(c,d))
print("Vectors a and b are orthonormal")
```

\textbf{Trace}

The \textbf{trace} of a square matrix is the sum of its diagonal entries:

$$tr(A) = \sum^{n}_{i = 0}A_{ii}$$

The trace has several algebraic properties:

\begin{itemize}
    \item $tr(A + B) = tr(A) + tr(B)$
    \item $tr(aA) = atr(A$)
    \item $tr(A^T) = tr(A)$
    \item $tr(AB) = tr(BA)$
\end{itemize}

Rhe trace of a matrix is equal to the sum of its eigenvalues:
$$tr(A) = \sum_iλ_i(A)$$

```python
# Using matrix A from the previous example
print(A)

print("\nTrace of matrix A: ", A.trace())

linalg.eigvals(A)

sum = 0

for x in linalg.eigvals(A):
    sum = sum + x

print("Sum of A's eigenvalues is: ", sum)
```

\textbf{Orthogonal Matrices} 

A matrix $Q \in R^{n*n}$ is said to be orthogonal if its columns are pairwise orthonormal. This means that:

$$Q^TQ = QQ^T = I$$

Orthogonal matrices have some nice properties:

\begin{itemize}
    \item They preserve inner products: $(Qx)^T(Qy) = x^TQ^TQy = x^Ty$ 
    \item They preserve 2-norms: $||Qx||_2 = \sqrt((Qx)^T(Qx)) = \sqrt(x^Tx) = ||x||_2$
\end{itemize}



```python
x = np.matrix([[1,4]])
y = np.matrix([[2,1]])
Q = np.matrix([[1, 0], [0, -1]])

print("Matrix Q: \n")
print(Q)

print("\nIs Q orthogonal? \n")
print("transpose(Q) = ")
print(Q.T)

print("\ntranspose(Q)*Q = ")
print(Q.T.dot(Q))

print("\nInner product transpose(x)y = ")
print(np.inner(x,y))

print("\n2-norm of x = ")
print(linalg.norm(x,2))

print("\n2-norm of Qx = ")
print(linalg.norm(Q*x.T,2))

```

$$A = UΣV^T$$

,where $U \in R^{m*m}$, $v \in R^{n*n}$ are orthogonal matrices and $Σ \in R^{m*n}$ a diagonal matrix with the singular values of A(denoted $σ_i$) on its diagonal.

\textbf{Pseudoinverses}

\begin{itemize}

If $A$ is invertible, then $A^+ = A^{-1}$. We can use SVD to compute the pseudoinverse of A:

$$A^+ = VΣ^+U^Τ$$

\textbf{Taylor's theorem}

Taylor's theorem has the ability to aproximate any single-input function to a Taylor series. <br>

If the function is continuously differentiable, we can compute the first order Taylor series. <br>
If the function is twice continuously differentiable, we can compute the second order Taylor series. <br>

And so on...  The more orders the Taylor series gets, the better the aproximation. <br>

We will illustrate the sin function along with a second order Taylor series approximation:
\begin{equation}
h(w) = g(w^0) + \frac{\mathrm{d}}{\mathrm{d}w}g(w^0)(w - w^0) +  \frac{1}{2}\frac{\mathrm{d}^2}{\mathrm{d}w^2}g(w^0)(w - w^0)^2.
\end{equation}
centered at the point $w^0 = 1$

```python
%matplotlib notebook
import autograd.numpy as np   
import matplotlib.pyplot as plt
from matplotlib import rcParams
from autograd import grad 
rcParams['figure.autolayout'] = True

# create function / first derivative
g = lambda w: np.sin(w) 

# create first and second derivative function
dgdw = grad(g)
dgdw2 = grad(dgdw)

# create first order taylor series approximation
# here w_0 is the center point of the expansion
first_order = lambda w_0,w: g(w_0) + dgdw(w_0)*(w - w_0)
 
# create second order taylor series approximation
# here w_0 is the center point of the expansion
second_order = lambda w_0,w: g(w_0) + dgdw(w_0)*(w - w_0) + 0.5*dgdw2(w_0)*(w - w_0)**2

# This code cell will not be shown in the HTML version of this notebook
# create area over which to evaluate everything
w = np.linspace(-5,5,200); w_0 = 1.0; w_=np.linspace(-2+w_0,2+w_0,200);

# define and evaluate the function, define derivative
gvals = [g(v) for v in w]

# create tangent line and quadratic
tangent = first_order(w_0,w_)
quadratic = second_order(w_0,w_)

# plot the function and derivative 
fig = plt.figure(figsize = (7,4))
plt.plot(w,gvals,c = 'k',linewidth=2,zorder = 1)
plt.plot(w_,tangent,c = [0,1,0.25],linewidth=2,zorder = 2)
plt.plot(w_,quadratic,c = [0,0.75,1],linewidth=2,zorder = 2)
plt.scatter(w_0,g(w_0),c = 'r',s=50,zorder = 3,edgecolor='k',linewidth=1)
plt.legend(['$sin(w)$','1st order series','2nd order series'],fontsize = 11)
plt.show()
```

With green we have the first order Taylor series approximation. <br>
With blue we have the second order Taylor series approximation.

\textbf{Basics}

A randomized experiment (e.g. a coin toss, die roll) has a fixed set of possible outcomes.<br>
This set is called the sample space and denoted $\Omega$ .

An event $A$ must produce outcomes that are possible to occur.<br>
So the event must be a set that contains values of the sample space.<br>

So $A \subset \Omega$  and $p(A) < p(\Omega)$

The Probabilty of an event is $0\leq p(A) \leq 1$

If $p(A) = 1$, we say that A occurs almost surely.<br>
If $p(A) = 0$, we say that A occurs almost never.

$p(\Omega) = 1$, because it contains all outcomes.<br>
$p(\emptyset) = 0$, because it contains no outcomes.<br>

let A be such that $P(A) = 0.4$ <br>

$p(A^{c}) = 1 - p(A) = 1 - 0.4 = 0.6$

\textbf{Union Rule}
$$ 
\begin{aligned}
    p (A \cup B) &= p(A) + p(B) - p(A \cap B)& \\
    &= p(A) + p(B) &(\textrm{if } A  \cap  B = \emptyset)
\end{aligned}
$$

\textbf{Boole's inequality}

also known as the union bound, says that for any finite or countable set of events, the probability that at least one of the events happens is no greater than the sum of the probabilities of the individual events:

$$
P(\bigcup_{n=1}^\infty B_n) = \sum_{n=1}^\infty P(B_n)
$$

\textbf{Conditional probability}

The conditional probability of event A given that event B has occurred is written P(A|B) and denoned as:

$$
    P(A|B)= \frac{P(A \cap B)}{ P(B)}
$$

\textbf{Chain rule}

comes immediately from the definition above

$$
    P(A \cap B) = P(A|B)P(B) = P(B|A)P(A)
$$

<br>

\textbf{Bayes' rule}


$$
    P(A|B)= \frac{P(B|A)P(A)}{ P(B)}
$$

Bayes rule provides us with a way to update our beliefs about a hypothesis $A$ based on the arrival of new, relevant piece of evidence $B$.<br>

The power of Bayes' rule is that in many situations where we want to compute $P(A|B)$ it turns out that it is difficult to do so directly, yet we might have direct information about $P(B|A)$. Bayes' rule enables us to compute $P(A|B)$ in terms of $P(B|A)$ .


For example, suppose that we are interested in diagnosing cancer in patients who visit a chest clinic.<br>

Let $A$ represent the event "Person has cancer".<br>
Let $B$ represent the event "Person is a smoker".<br>

We know the probability of the prior event $P(A)=0.1$ on the basis of past data <br>
(10% of patients entering the clinic turn out to have cancer).<br> 

We want to compute the probability of the posterior event $P(A|B)$. It is difficult to find this out directly. However,<br>
we are likely to know $P(B)$ by considering the percentage of patients who smoke – suppose $P(B)=0.5$ . 

We are also likely to know $P(B|A)$ by checking from our record the proportion of smokers among those diagnosed.<br> 
Suppose $P(B|A)=0.8$ .

We can now use Bayes' rule to compute:

 $$P(A|B) = (0.8 * 0.1)/0.5 = 0.16$$

Thus, in the light of evidence that the person is a smoker we revise our prior probability from 0.1 to a posterior probability of 0.16 . This is a significance increase, but it is still unlikely that the person has cancer.


This  means that we can utilize prior knowledge to improve our probability estimations.

\textbf{Covariance}

Covariance is a measure of the linear relationship between two random variables.

We calculate covariance with numpy.cov(x)  <br>
If we use a matrix instead of vector it calculates the covariance matrix instead.

The Correlation of two vectors is calculated with numpy.corrcoef(x,y)

\textbf{Random vectors}

Univariate distributions, are distributions of single variables.<br>
But we can also have multivariate distributions which give distributions of random vectors:<br>


The covariance matrix we talked about earlier is the generalization of Variance and Covariance <br>
from univariate distributions to multivariate distributions. It is calculated as:

![title](img/covariance_matrix.png)

```python
import numpy as np
np.set_printoptions(precision=3)

x = [2.1, 1.2,  5.2]
y = [-3,  2.1,  -1.12]

print('Expexted value of x is: %.3f' %np.mean(x) ,'\n')

print('var(x) is: %.3f'%np.var(x, axis=0, ddof=1) ,'\n')
print('var(y) is: %.3f'%np.var(y, axis=0, ddof=1) ,'\n')

X = np.stack((x, y), axis=0)

print('Covariance matrix of X is: \n ' , np.cov(X) ,'\n')

print('Correlation matrix of X is: \n' , np.corrcoef(X) ,'\n')
```

Notice that the diagonal elements of cov(X) are the var(x) and var(y) of the respective vectors.


\textbf{The Gaussian distribution}

The code below shows that with numpy.random.normal we can generate random variables that follow the normal distribution<br>
with given mean and standard diviation.

```python
import matplotlib.pyplot as plt
import numpy as np

mean, sigma = 0.5, 0.1
s = np.random.normal(mean, sigma, 1000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mean)**2 / (2 * sigma**2) ),       linewidth=3, color='y')
plt.show()
```
